```python
surveys = pd.read_csv("surveys.csv")
surveys[pd.isnull(surveys["weight"])]["weight"]
import matplotlib
%matplotlib inline
```

```python
data = surveys[(surveys["sex"].isin(["F", "M"])) & (surveys.weight > 0)]
h = data.groupby(["plot_id", "sex"]).mean()["weight"].unstack()
h.plot(kind = "bar", stacked = True)
```
