```python
plot_names = list(pd.unique(surveys["species_id"]))
print(len(plot_names))
print(surveys["plot_id"].nunique())
```

```python
sorted_by_sex = surveys.groupby("sex")
sorted_by_sex.count()
```

```python
species = surveys.groupby("species_id").count()["record_id"]
species
```

```python
import matplotlib
%matplotlib inline
```

```python
species.plot(kind="bar")
```

```python
species_weight = surveys.groupby("species_id").mean()["weight"]
species_weight.plot(kind = "bar")
```

```python
num_sex = surveys.groupby("sex").count()["record_id"]
num_sex.plot(kind = "bar")
```

```python
plot_sex = surveys.groupby(["plot_id", "sex"]).sum()["weight"]
plot_sex = pd.DataFrame(plot_sex).unstack()
plot_sex.plot(kind = "bar", stacked = True)
```
